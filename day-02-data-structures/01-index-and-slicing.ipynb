slicing (dilimlemek)

```python
name = "Ömer"
surname = "Ocak "
fullName = name + " " + surname 
print(fullName)
```

```python
fullName * 5 
```

```python
barcode = "ABC21FHN46346"
```

```python
barcode[0] + barcode[1] + barcode[2]
```

```python
# slicing(dilimlemek), starting index, stopping index, stepping size
```

```python
barcode[::] # tümünü al
```

```python
# barcode[starting index : stopping index : stepping size ]  
```

```python
barcode[3::] # ilk 3 elemanı eledi
```

```python
barcode[:3:] # 3. index te dur ama alma
```

```python
barcode[3:5]
```

```python
barcode[::-1] # tersten yazma
```

```python
integers = "0123456789"
```

```python
integers[0::2] # çift sayılar
```

```python
integers[1::2] # tek sayılar
```

```python
integers[0::3] # 3 ün katları
```

```python
integers[::-1]
```
